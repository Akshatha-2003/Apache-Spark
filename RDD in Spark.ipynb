```python
# start a spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Basics').getOrCreate()
```

```python
list1 = [("Alice",20) , ("Ram",45) , ("Slice",54) , ("Yamuna",76) , ("Alice",52)]
rdd_list = spark.sparkContext.parallelize(list1)
```

```python
# print the rdd values
rdd_list.collect()
```

***Transformations in RDD***

**MAP - Convert all the names into upper case**

```python
rdd_mapped = rdd_list.map(lambda x: (x[0].upper(),x[1]))
```

***Transformations are lazy evaluation , they dont execute immediately . So invoke an action called **collect()** to print all results to the console.***

```python
rdd_mapped.collect()
```

**FILTER - Filter only those records where marks are greater than 30**

```python
rdd_filtered = rdd_list.filter(lambda x:x[1]>30)
```

```python
rdd_filtered.collect()
```

**ReducebyKey - Find total marks for each student**

```python
rdd_keyby = rdd_list.reduceByKey(lambda x, y:x+y)
```

```python
rdd_keyby .collect()
```

**Sortby - Sort the marks in desc order**

```python
rdd_sorted = rdd_list.sortBy(lambda x:x[1], ascending = False)
```

```python
rdd_sorted .collect()
```

**FlatMap - Flattens the result into a single list**

```python
rdd_flatten = rdd_list.flatMap(lambda x: x[0].split(" "))
```

```python
rdd_flatten.collect()
```

***Actions in RDD***

**collect() - print the result in console**

**count () - count of elements in rdd**

```python
rdd_list.count()
```

**first() - first element of RDD**

```python
rdd_list.first()
```

**take() - limits the output to only n elements**

```python
rdd_list.take(2)
```

**foreach() - Prints every elements of RDD**

```python
rdd_list.foreach(lambda x:print(x))
```

***Why the above code returns None?***

**Because foreach runs on the executor and not on the driver . The printing of elemeents happens on your worker node , not on driver .**

***BELOW WAYS CAN BE SIMPLY USED TO PRINT EVERY ELEMENT OF RDD***

```python
for x in rdd_list.collect():
  print(x)
```

```python
rdd_list.collect()
```

***When to Actually Use foreach()***

*foreach() is useful when you want to write data somewhere without bringing it to Driver.*

***Save RDD as a text file***

***Read from a text file***

```python
rdd_text = spark.sparkContext.textFile("output.txt")
```

```python
rdd_text.collect()
```
